Please recreate the bar chart seen below.

# Use this data to generate the plot!
* resorts = ["Crested Butte", "Aspen", "Telluride", "Wolf Creek"]
* snow_last_24 = [29,15,15,23]
* snow_next_24 = [16,6,19,30]

```python
resorts = ["Crested Butte", "Aspen", "Telluride", "Wolf Creek"]
snow_last_24 = [29,15,15,23]
snow_next_24 = [16,6,19,30]
```

```python
plt.figure(figsize=(9,6))
plt.bar(resorts, snow_last_24, width = 0.4, color= '#7ed6df', label = 'Last 24hrs')
plt.bar(resorts, snow_next_24,bottom=snow_last_24, width = 0.4, color= '#74b9ff', label = 'Next 24hrs')
plt.title('Co Snow Storm Report')
plt.ylabel('Inches')
plt.yticks([0,10,20,30,40,50])
plt.legend(shadow = True, facecolor= 'white', loc= 2)
```

Recreate the pie chart seen below:

* I've provided lists for the actual numbers, the labels, and the colors

* Notice the percentages in each slice!

```python
# Use these lists to generate the chart!
labels = [
    "Building Construction", 
    "Deferred Maintenance",
    "General Contractor Fees",
    "Owner Costs",
    "Site Construction"
]
costs = [
    47945975,
    3753584,
    7340612,
    18247137,
    5766685
]
colors = ["#355070", "#6d597a", "#b56576", "#e56b6f", "#eaac8b"]
```

```python
import pandas as pd
```

```python
houses = pd.read_csv('DataAnalysis/data/kc_house_data.csv', parse_dates=['date'])
```

```python
week_day = houses.date.dt.dayofweek.value_counts()
plt.bar(week_day.index.values, week_day)
plt.xticks([0,1,2,3,4,5,6], ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
plt.title('Sales By Week')
```

```python
month_day = houses.date.dt.month.value_counts().sort_index()
month_day
```

```python
plt.plot(month_day.index.values, month_day)
plt.title('Sales By Month')
plt.xlabel('Month')
plt.ylabel('Home Sales')
```
