```python
salaries = pd.read_csv('DataAnalysis/data/Salaries.csv', low_memory=False)
```

```python
fig, axs = plt.subplots(1,2)
plt.figure(figsize=(10,10))
ufos.year.value_counts().sort_index().plot(ax = axs[0])
axs[0].set_title('Sighting By Year')
ufos.Month.value_counts().sort_index().plot(kind='bar', ax = axs[1])
axs[1].set_title('Sighting By Month')
```

```python
fig, axs = plt.subplots(2,3, figsize = (14,8))
fig.suptitle('Ufo Sightings By Month', fontsize= 14)
ufos[ufos.year == 2014].Month.value_counts().sort_index().rename(month).plot(kind='bar', color = '#1abc9c', ax = axs[0][0], title = '2014')
ufos[ufos.year == 2015].Month.value_counts().sort_index().rename(month).plot(kind='bar', color = '#3498db', ax = axs[0][1], title = '2015')
ufos[ufos.year == 2016].Month.value_counts().sort_index().rename(month).plot(kind='bar', color = '#9b59b6', ax = axs[0][2], title = '2016')
ufos[ufos.year == 2017].Month.value_counts().sort_index().rename(month).plot(kind='bar', color = '#f1c40f', ax = axs[1][0], title = '2017')
ufos[ufos.year == 2018].Month.value_counts().sort_index().rename(month).plot(kind='bar', color = '#30336b', ax = axs[1][1], title = '2018')
ufos[ufos.year == 2019].Month.value_counts().sort_index().rename(month).plot(kind='bar', color = '#6ab04c', ax = axs[1][2], title = '2019')
plt.tight_layout()
```
