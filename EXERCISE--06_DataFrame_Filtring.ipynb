# Filtering Data Exercise
Answer the following questions using the `bestsellers.csv` dataset

### Find the books written by Pete Souza

```python
import pandas as pd
```

```python
sellers = pd.read_csv('DataAnalysis/data/bestsellers.csv')
sellers
```

```python
sellers[sellers.Author == 'Pete Souza']
```

### Find the books that are under 10 dollars

```python
sellers[sellers.Price < 10]
```

#### Find the books that have a price between 50 and 60 dollars

```python
sellers[sellers.Price.between(50, 60)]
```

#### Find all the books written by Kristin Hannah, or Weir, or Delia Owens

```python
kristian = sellers.Author == 'Kristin Hannah'
andy = sellers.Author == 'Andy Weir'
delia = sellers.Author == 'Delia Owens'
```

```python
sellers[kristian | andy | delia]
```

```python
sellers[kristian]
```

```python
sellers[andy]
```

#### Find the Non Fiction books that are rated 4.9

```python
gn_isna = sellers.Genre == 'Non Fiction'
rating = sellers['User Rating'] == 4.9
sellers[gn_isna & rating]
```

#### Find the fiction book with the lowest User Rating

```python
sellers[sellers.Genre == 'Fiction'].sort_values('User Rating').head()
```

```python
lowest = sellers['User Rating'] == 3.3
fin_gn = sellers.Genre == 'Fiction'
sellers[fin_gn & lowest]
```

```python
sellers[lowest]
```

#### Find 2012's top 5 Fiction books with the most Reviews

```python
df = sellers[(sellers.Year == 2012) & (sellers.Genre == 'Fiction')]
df.sort_values('Reviews', ascending=False).head()
```

### Create a bar plot showing the 5 authors who have the most books with a rating under 4.5

```python
top_author = sellers[sellers['User Rating'] < 4.5].Author.value_counts().head()
top_author.plot(kind='bar')
```
